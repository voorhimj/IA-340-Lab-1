```python

import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the dataframe is named df
df = pd.read_sql(sql,conn)

# Calculate the year over year income growth for each state
df.sort_values(["name", "year"], inplace=True)
df["growth"] = df.groupby("name")["income"].pct_change()

# Pivot the data so each state is a column and each row is a year
pivot_df = df.pivot(index='year', columns='name', values='growth')

# Plot the data
plt.figure(figsize=(20,10))
for column in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[column], marker='', linewidth=1, alpha=0.9, label=column)

plt.legend(loc=2, ncol=2)
plt.title("Year Over Year Income Growth For Each State", loc='left', fontsize=12, fontweight=0, color='orange')
plt.xlabel("Year")
plt.ylabel("Growth")
plt.show()


#This script calculates the year over year income growth for each state, then plots the results on a line chart. The x-axis represents the year and the y-axis represents the growth. Each line in the chart represents a state, with the color and label indicating which state it corresponds to.
```
